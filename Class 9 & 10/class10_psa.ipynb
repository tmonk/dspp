# Non-linear machine learning - example non-linear functions

We now move on to better - yet more difficult - prediction.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
vi = pd.read_csv('vi.csv')
vi['timestamp'] = pd.to_datetime(vi['Date'],format="%b %y")

plt.scatter(vi['timestamp'],vi['Labour'], color='r')
plt.scatter(vi['timestamp'],vi['Conservatives'], color='b')
plt.scatter(vi['timestamp'],vi['Liberal Democrats'], color='y')
```

```python
import datetime
# Con less Lab
vi['con-lab'] = (vi['Conservatives'] - vi['Labour'])
vi.dropna(subset='con-lab')
vi['numeric-date'] = (vi['timestamp'] - datetime.datetime(2017,7,1)).astype('timedelta64[D]')
print(vi['numeric-date'])
plt.scatter(vi['timestamp'],vi['con-lab'], color='k')
```

```python
# Low order polynomial
order = 2
x = vi['numeric-date']
y = vi['con-lab']
z = np.polyfit(x, y, order)
xp = np.linspace(min(vi['numeric-date']), max(vi['numeric-date']), 100)
p = np.poly1d(z)
_ = plt.plot(x, y, '.',color='k')
_ = plt.plot(xp, p(xp),color='b')
```

```python
# High order polynomial
order = 15
x = vi['numeric-date']
y = vi['con-lab']
z = np.polyfit(x, y, order)
xp = np.linspace(min(vi['numeric-date']), max(vi['numeric-date']), 100)
p = np.poly1d(z)
_ = plt.plot(x, y, '.',color='k')
_ = plt.plot(xp, p(xp),color='b')
```
